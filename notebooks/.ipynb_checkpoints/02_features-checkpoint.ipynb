```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/raw/ai4i2020.csv')
print(f"Shape: {df.shape}")
print("Data loaded!")
```

```python
df['rolling_mean_temp'] = df['Air temperature [K]'].rolling(window=10).mean()
df['rolling_mean_torque'] = df['Torque [Nm]'].rolling(window=10).mean()
df['rolling_mean_speed'] = df['Rotational speed [rpm]'].rolling(window=10).mean()

plt.figure(figsize=(12,4))
plt.plot(df['Air temperature [K]'][:200], label='Original', alpha=0.7)
plt.plot(df['rolling_mean_temp'][:200], label='Rolling Mean', linewidth=2)
plt.title('Rolling Mean - Air Temperature')
plt.legend()
plt.savefig('../reports/figures/06_rolling_mean.png')
plt.show()
print("Rolling mean calculated!")
```

```python
df['rolling_std_temp'] = df['Air temperature [K]'].rolling(window=10).std()
df['rolling_std_torque'] = df['Torque [Nm]'].rolling(window=10).std()

plt.figure(figsize=(12,5))
plt.plot(df['rolling_std_temp'][:200], label='Rolling Std', color='red')
plt.title('Rolling Standard Deviation - Air Temperature')
plt.legend()
plt.savefig('../reports/figures/07_rolling_std.png')
plt.show()
print("Rolling std calculated!")
```

```python
df['signal_variance'] = df['Torque [Nm]'].rolling(window=10).var()

plt.figure(figsize=(12,4))
plt.plot(df['signal_variance'][:200], label='Signal Variance', color='green')
plt.title('Signal Variance - Torque')
plt.legend()
plt.savefig('../reports/figures/08_signal_variance.png')
plt.show()
print("Signal variance calculated!")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Air temperature [K]'], df['Process temperature [K]'], 
            c=df['Machine failure'], cmap='coolwarm', alpha=0.5)
plt.colorbar(label='Machine Failure')
plt.title('Air vs Process Temperature')
plt.xlabel('Air Temperature [K]')
plt.ylabel('Process Temperature [K]')
plt.savefig('../reports/figures/06_temperature_scatter.png')
plt.show()
```

```python
colors = ['red' if x==1 else 'blue' for x in df['Machine failure']]
plt.figure(figsize=(10,6))
plt.scatter(df['Rotational speed [rpm]'], df['Torque [Nm]'],
            c=colors, alpha=0.3)
plt.title('Rotational Speed vs Torque')
plt.xlabel('Rotational Speed [rpm]')
plt.ylabel('Torque [Nm]')
plt.savefig('../reports/figures/07_speed_vs_torque.png')
plt.show()
```

```python
df['temp_diff'] = df['Process temperature [K]'] - df['Air temperature [K]']
fig, axes = plt.subplots(1, 2, figsize=(12,4))
df[df['Machine failure']==0]['temp_diff'].hist(bins=30, ax=axes[0], color='blue')
axes[0].set_title('Temp Difference - Normal')
df[df['Machine failure']==1]['temp_diff'].hist(bins=30, ax=axes[1], color='red')
axes[1].set_title('Temp Difference - Failure')
plt.savefig('../reports/figures/08_temp_difference.png')
plt.show()
```

```python
failure_cols = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']
failure_counts = df[failure_cols].sum().sort_values(ascending=False)
plt.figure(figsize=(8,5))
failure_counts.plot(kind='bar', color=['red','blue','orange','green','gray'])
plt.title('Failure Types Distribution')
plt.xlabel('Failure Type')
plt.ylabel('Count')
plt.savefig('../reports/figures/09_failure_by_type.png')
plt.show()
```
